### ✅ المهمة 1.3 — الجمع `__add__` و `__radd__`

قواعد الجمع:
- **شعاع + شعاع:** جمع عنصر بعنصر (يجب أن يكون الطولان متساويين)
- **شعاع + رقم:** أضف الرقم لكل عنصر
- **رقم + شعاع:** نفس النتيجة (هذا هو `__radd__`)

````python
    def __add__(self, other):
        if isinstance(other, Sho3a3):
            assert len(self) == len(other)
            return Sho3a3([a + b for a, b in zip(self, other)])

        else:
            return Sho3a3([a + other for a in self])

    def __radd__(self, other):
        return self.__add__(other)
````

### ✅ المهمة 1.4 — الطرح `__sub__` و `__rsub__`

نفس قواعد الجمع لكن بالطرح:

````python
def __sub__(self, other):
        if isinstance(other, Sho3a3):
            assert len(self) == len(other)
            return Sho3a3([a - b for a, b in zip(self, other)])
        else:
            return Sho3a3([a - other for a in self])

def __rsub__(self, other):
        return Sho3a3([other - a for a in self])
````

> ⚠️ **تنبيه:** الترتيب مهم في الطرح! `other - self` ≠ `self - other`

**اختبار:**

````python
a = Sho3a3([5, 6, 7])
b = Sho3a3([1, 2, 3])
print(a - b)    # sho3a3:[4, 4, 4]
print(a - 1)    # sho3a3:[4, 5, 6]
print(10 - a)   # sho3a3:[5, 4, 3]
````

### ✅ المهمة 1.5 — الضرب `__mul__` و `__rmul__`

للضرب، لدينا حالتان:
- **شعاع × شعاع:** الجداء النقطي (dot product) — يُرجع **رقماً واحداً**
- **شعاع × رقم:** ضرب كل عنصر بالرقم — يُرجع **شعاعاً جديداً**

**الجداء النقطي:**
$$\vec{a} \cdot \vec{b} = a_1 b_1 + a_2 b_2 + a_3 b_3 + \ldots$$

````python
def __mul__(self, other):
        if isinstance(other, Sho3a3):
            assert len(self) == len(other)
            return sum(a * b for a, b in zip(self, other))

        else:
            return Sho3a3([a * other for a in self])

def __rmul__(self, other):
        return self.__mul__(other)
````

**اختبار:**

````python
a = Sho3a3([1, 2, 3])
b = Sho3a3([4, 5, 6])
print(a * b)    # 32  (1×4 + 2×5 + 3×6 = 4+10+18)
print(a * 2)    # sho3a3:[2, 4, 6]
print(3 * a)    # sho3a3:[3, 6, 9]
````

## 🏗️ الجزء الثاني: فئة `Masfofah` (المصفوفة)

> الملف: `masfofah.py`

المصفوفة عبارة عن قائمة من الشعاعات. كل صف هو كائن `Sho3a3`.

````python
%%writefile masfofah.py
from sho3a3 import Sho3a3
from masfofah import Masfofah

class Masfofah:
    def __init__(self, data):
        self.rows = [row if isinstance(row, Sho3a3) else Sho3a3(row) for row in data]
        self.shaper = (len(self.rows), len(self.rows[0]) if self.rows else 0)

    def __repr__(self):
        return "masfofah:\n" + "\n".join(str(r) for r in self.rows)

    def shape(self):
        return self.shaper

    def __len__(self):
        return len(self.rows)

    def __getitem__(self, place):
        return self.rows[place]

    def __add__(self, other):
        assert self.shape() == other.shape()
        return Masfofah([self.rows[i] + other.rows[i] for i in range(len(self.rows))])

    def __sub__(self, other):
        assert self.shape() == other.shape()
        return Masfofah([self.rows[i] - other.rows[i] for i in range(len(self.rows))])

    def extract_columns(self):
        r, c = self.shape()
        cols = []
        for j in range(c):
            cols.append(Sho3a3([self.rows[i][j] for i in range(r)]))
        return Masfofah(cols)

    def __mul__(self, other):
        assert self.shape()[1] == other.shape()[0]
        other_cols = other.extract_columns()
        res = []
        for row in self.rows:
            res.append([row * col for col in other_cols.rows])
        return Masfofah(res)

    def man8ool(self):
        return self.extract_columns()

    def sum(self, axis: int):
        if axis == 0:
            return Sho3a3([c.sum() for c in self.extract_columns().rows])
        return Sho3a3([r.sum() for r in self.rows])

    def add(self, other, axis=0):
        if axis == 1:
            self.rows.extend(other.rows)
        else:
            for i in range(len(self.rows)):
                self.rows[i].sho3a3.extend(other.rows[i].sho3a3)
        self.shaper = (len(self.rows), len(self.rows[0]))
````

### ✅ المهمة 2.4 — ضرب المصفوفات `__mul__`

ضرب المصفوفات يعمل هكذا:

$$C[i][j] = \text{(الصف } i \text{ من } A) \cdot \text{(العمود } j \text{ من } B)$$

**الشرط:** عدد أعمدة `A` يجب أن يساوي عدد صفوف `B`

```
A (2×3)  ×  B (3×2)  =  C (2×2)
```

````python
def __mul__(self, other):
        assert self.shape()[1] == other.shape()[0]
        other_cols = other.extract_columns()
        res = []
        for row in self.rows:
            res.append([row * col for col in other_cols.rows])
        return Masfofah(res)
````

> 💡 **تلميح:** استخدم `extract_columns()` لاستخراج أعمدة `other`، ثم استخدم `*` من `Sho3a3` للجداء النقطي.

**اختبار:**

````python
a = Masfofah([[1, 2], [3, 4]])
b = Masfofah([[5, 6], [7, 8]])
c = a * b
print(c)
# masfofah:
# sho3a3:[19, 22]
# sho3a3:[43, 50]
````

### ✅ المهمة 3.2 — مصفوفة الوحدة (Identity Matrix)

مصفوفة الوحدة هي مصفوفة مربعة فيها `1` على القطر الرئيسي و `0` في بقية العناصر:

```
3×3 Identity:
1  0  0
0  1  0
0  0  1
```

````python
def identity_masfofah(n):
    return Masfofah([[1 if i == j else 0 for j in range(n)] for i in range(n)])
````

> 💡 **تلميح:** ابدأ بمصفوفة أصفار، ثم اجعل `[i][i] = 1` لكل `i`.

### 🌟 المهمة 3.4 (تحدي اختياري) — مولّد الأعداد العشوائية

بدلاً من استيراد `random` من Python، اكتب مولّد أعداد عشوائية بسيط خاصتك!

يمكنك استخدام **خوارزمية Linear Congruential Generator (LCG):**

$$X_{n+1} = (a \times X_n + c) \mod m$$

````python
class Generator:
    def __init__(self, seed=42):
        self.state = seed
        self.a = 1664525
        self.c = 1013904223
        self.m = 2**32

    def next_int(self) -> int:
        self.state = (self.a * self.state + self.c) % self.m
        return self.state

    def random_norm(self) -> float:
        return self.next_int() / self.m

    def random_int(self, low=0, high=100) -> int:
        return low + (self.next_int() % (high - low + 1))
````

````python
def random_matrix(rows, columns):
    g = Generator()
    data = []

    for i in range(rows):
        row = []
        for j in range(columns):
            row.append(g.random_norm())
        data.append(row)

    return Masfofah(data)
````

````python
print(random_matrix(2, 3))
````
